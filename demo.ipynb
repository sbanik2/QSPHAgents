```python
from qsph_agent.utils import load_keys_and_config
from qsph_agent.rag_data import build_database_if_needed, load_and_split_database
from qsph_agent.agent_workflow import (
    build_predictor,
    build_qsph_graph,
    make_initial_state,
    summarize_final_state,
)

from openai import OpenAI
import numpy as np
from pprint import pprint


# --------------------------------------------------
# 1) Load keys + config
# --------------------------------------------------
MP_API_KEY, CONFIG = load_keys_and_config("config.yaml")

# --------------------------------------------------
# 2) Build / load database
# --------------------------------------------------
datafile = build_database_if_needed(MP_API_KEY, CONFIG)
train_set, test_set, full_data = load_and_split_database(datafile)

print(f"Train size: {len(train_set)}, Test size: {len(test_set)}")

# --------------------------------------------------
# 3) Build predictor + graph
# --------------------------------------------------
predictor = build_predictor(train_set, n_neighbors=6)
graph = build_qsph_graph()

# --------------------------------------------------
# 4) (Optional) Run one example and print human summary
# --------------------------------------------------
first_key = list(test_set.keys())[0]
first_entry = test_set[first_key]
state = make_initial_state(first_entry, train_set, predictor, top_k=7)
final_state = graph.invoke(state)
#summarize_final_state(final_state)


# --------------------------------------------------
# 5) Helper functions for evaluation
# --------------------------------------------------
def normalize_shape(s):
    if not s:
        return ""
    s = s.lower()
    if "u-shaped" in s:
        return "u-shaped"
    if "v-shaped" in s:
        return "v-shaped"
    if "flat" in s:
        return "flat"
    if "pseudogap" in s:
        return "pseudogap"
    return s.strip()


def normalize_asym(s):
    if not s:
        return ""
    s = s.lower()
    if "valence" in s and "higher" in s:
        return "valence>conduction"
    if "conduction" in s and "higher" in s:
        return "conduction>valence"
    return "other"


def safe_mean(xs):
    return float(np.mean(xs)) if xs else float("nan")


# ---- semantic similarity via OpenAI embeddings ----
client = OpenAI()  # uses OPENAI_API_KEY already set by load_keys_and_config


def cosine_sim(a, b):
    a = np.array(a, dtype=float)
    b = np.array(b, dtype=float)
    denom = np.linalg.norm(a) * np.linalg.norm(b)
    if denom == 0:
        return 0.0
    return float(a @ b / denom)


def semantic_similarity(a: str, b: str) -> float:
    a = (a or "").strip()
    b = (b or "").strip()
    if not a or not b:
        return 0.0
    resp = client.embeddings.create(
        model="text-embedding-3-small",
        input=[a, b],
    )
    v1 = resp.data[0].embedding
    v2 = resp.data[1].embedding
    return cosine_sim(v1, v2)


# --------------------------------------------------
# 6) Loop over ALL test entries and collect metrics
# --------------------------------------------------
cls_correct = []       # material_classification exact match
shape_correct = []     # overall_dos_shape (normalized) exact match
asym_correct = []      # asymmetry_comment (normalized) exact match

shape_sims = []        # semantic similarity for shape sentence
asym_sims = []         # semantic similarity for asymmetry sentence

rows = []              # per-sample inspection


for key, test_entry in test_set.items():
    # run full agentic loop
    state = make_initial_state(test_entry, train_set, predictor, top_k=7)
    final_state = graph.invoke(state)

    # ground truth dict from database
    gt_dict = test_entry["dos"]["dos_description_dict"]

    # structured hypothesis from agent (set in generation_node)
    hypo_struct = final_state.get("prev_hypothesis_struct")
    if not hypo_struct:
        # if for some reason we didn't get structured_response, skip
        continue

    # ------------ discrete labels ------------
    gt_cls = gt_dict.get("material_classification")
    pred_cls = hypo_struct.get("material_classification")

    gt_shape_norm = normalize_shape(gt_dict.get("overall_dos_shape"))
    pred_shape_norm = normalize_shape(hypo_struct.get("overall_dos_shape"))

    gt_asym_norm = normalize_asym(gt_dict.get("asymmetry_comment"))
    pred_asym_norm = normalize_asym(hypo_struct.get("asymmetry_comment"))

    if gt_cls and pred_cls:
        cls_correct.append(int(gt_cls == pred_cls))

    if gt_shape_norm and pred_shape_norm:
        shape_correct.append(int(gt_shape_norm == pred_shape_norm))

    if gt_asym_norm and pred_asym_norm:
        asym_correct.append(int(gt_asym_norm == pred_asym_norm))

    # ------------ semantic similarity on raw text ------------
    raw_gt_shape = gt_dict.get("overall_dos_shape") or ""
    raw_pred_shape = hypo_struct.get("overall_dos_shape") or ""

    raw_gt_asym = gt_dict.get("asymmetry_comment") or ""
    raw_pred_asym = hypo_struct.get("asymmetry_comment") or ""

    shape_sim = semantic_similarity(raw_gt_shape, raw_pred_shape)
    shape_sims.append(shape_sim)

    if raw_gt_asym and raw_pred_asym:
        asym_sim = semantic_similarity(raw_gt_asym, raw_pred_asym)
        asym_sims.append(asym_sim)
    else:
        asym_sim = None

    rows.append({
        "key": key,
        "gt_cls": gt_cls,
        "pred_cls": pred_cls,
        "gt_shape": gt_shape_norm,
        "pred_shape": pred_shape_norm,
        "gt_asym": gt_asym_norm,
        "pred_asym": pred_asym_norm,
        "shape_sim": shape_sim,
        "asym_sim": asym_sim,
    })


# --------------------------------------------------
# 7) Print global metrics
# --------------------------------------------------
print("\n=== Discrete accuracies ===")
print("material_classification accuracy:", safe_mean(cls_correct))
print("overall_dos_shape accuracy:", safe_mean(shape_correct))
print("asymmetry_comment accuracy:", safe_mean(asym_correct))

print("\n=== Semantic similarities (cosine on embeddings) ===")
print("overall_dos_shape mean cosine sim:", safe_mean(shape_sims))
print("asymmetry_comment mean cosine sim:", safe_mean(asym_sims))

print("\nSample per-material rows:")
pprint(rows[:3])
```

```python


# ---- Data from your three runs ----
material_systems = ["Si", "BN", "GaN"]

# Train + Test sizes
train_sizes = np.array([20, 13, 3])
test_sizes = np.array([5, 4, 1])

# Metrics
material_classification_acc = np.array([1.0, 0.5, 1.0])
overall_dos_shape_cosine = np.array([
    0.545804450847194,
    0.48406803243313024,
    0.5893941618245648
])

# ---- Plot ----
x = np.arange(len(material_systems))
width = 0.3  # bar width

fig, ax1 = plt.subplots(figsize=(7, 2.5))

# Bars: classification accuracy and cosine similarity
bars_acc = ax1.bar(
    x - width/2,
    material_classification_acc,
    width,
    label="Material Classification Accuracy",
    color="#1f77b4",
    alpha=0.85,
)

bars_cos = ax1.bar(
    x + width/2,
    overall_dos_shape_cosine,
    width,
    label="Overall DOS Shape (Mean Cosine Sim)",
    color="#ff7f0e",
    alpha=0.85,
)

ax1.set_ylim(0, 1.70)
ax1.set_ylabel("Score (0â€“1)")

# --- Add train/test info under each tick ---
xtick_labels = [
    f"{name}\n(ref Data={tr}, test={te})"
    for name, tr, te in zip(material_systems, train_sizes, test_sizes)
]

ax1.set_xticks(x)
ax1.set_xticklabels(xtick_labels)

# Legend styling
ax1.legend(loc="upper right", frameon=False)

# ---- Annotate bar values ----
def autolabel(bars, fmt="{:.2f}", fontsize=16):
    for bar in bars:
        height = bar.get_height()
        ax1.annotate(
            fmt.format(height),
            xy=(bar.get_x() + bar.get_width() / 2, height),
            xytext=(0, 3),
            textcoords="offset points",
            ha="center",
            va="bottom",
            fontsize=fontsize,
        )

autolabel(bars_acc, fontsize=10)
autolabel(bars_cos, fontsize=10)

# Optional: smaller tick size on y-axis
ax1.tick_params(axis='y', labelsize=12)

plt.savefig("analysis.png",dpi=300,bbox_inches='tight')
```
